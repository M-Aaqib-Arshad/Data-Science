# Finding and Cleaning the values

```python
ship.isnull().sum()
```

```python
# del the deck column
newShip = ship.drop(['deck'],axis=1)
```

```python
# Remove the missing values
newShip = newShip.dropna()
newShip.isnull().sum()
```

```python
# Counts the value 
newShip["sex"].value_counts()
```

```python
ship.describe()
```

```python
newShip.describe()
```

```python
newShip.columns
```

```python
sns.boxplot(x='sex',y='age',data=newShip)
# To check outliers
```

```python
# To check outliers 
sns.displot(newShip['age'])
```

```python
# removing Outliers 
newShip = newShip[newShip['age']<65]
newShip['age'].max()
```

```python
sns.boxplot(x='sex',y='age',data=newShip)
```

```python
sns.distplot(newShip['age'])
```

```python
# remove the child to make normalize the bel curve
newShip = newShip[newShip['age']>10]
```

```python
sns.distplot(newShip['age'])
```

```python
newShip.boxplot()
```

```python
newShip = newShip[newShip['fare']<300]
```

```python
newShip.boxplot()
```

```python
sns.distplot(newShip['fare'])
```

```python
newShip.hist()
```

```python
sns.barplot(x="sex",y="survived",data=newShip,hue='who')
```

```python
pd.value_counts(newShip['survived']).plot.bar()
```

```python
# ks = newShip.groupby(['sex','survived'])
# ks.mean()
newShip.groupby(['sex', 'survived','pclass'] )['survived'].sum()
```

```python
grouped_data = newShip.groupby(['sex']).sum()
grouped_data.plot(kind='bar')
plt.xlabel('sex')
plt.ylabel('Total Sum')
plt.title('Sum of Values by Sex')

# Show the plot

plt.show()
```
